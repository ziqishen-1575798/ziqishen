```python
read_higgs = pd.read_pickle('higgs_100000_pt_1000_1200.pkl')
read_qcd = pd.read_pickle('qcd_100000_pt_1000_1200.pkl')
read_qcd.describe()
```

```python
infile = open ("higgs_100000_pt_1000_1200.pkl",'rb')
higgs = pickle.load(infile)

infile = open ("qcd_100000_pt_1000_1200.pkl",'rb')
qcd = pickle.load(infile)

q_mass=qcd['mass']
h_mass=higgs['mass']
plt.hist(q_mass,  bins = 50, density = True, label = 'QCD')
plt.hist(h_mass,  bins = 50, density = True, label = 'Higgs', alpha = 0.7)
plt.title('Stacked Mass Histogram for Higgs and QCD')
plt.xlim(0,500)
plt.xlabel('Mass')
plt.legend()
plt.show()
```

This histogram shows that the Higgs boson is likely to have a mass of 125 GeV/c^2, and with a narrow standard deviation compared to the QCD background. This tells us that the Higgs boson is more massive compare to the background particles.
